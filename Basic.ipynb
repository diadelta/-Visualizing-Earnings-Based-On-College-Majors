# Do students that majored in subjects that were majority female make more money?

```python
recent_grads_cols = recent_grads[["Men", "Women", "Median"]]
recent_grads_cols = recent_grads_cols.reset_index()
recent_grads_cols
```

```python
women_num = recent_grads_cols["Women"].values
men_num = recent_grads_cols["Men"].values

w_median = []
m_median = []

for i in range(0,len(women_num)):
    if women_num[i] > men_num[i]:
        w_median.append(recent_grads_cols.loc[i]["Median"])
    else:
        m_median.append(recent_grads_cols.loc[i]["Median"])
```

```python
m_median_avg = sum(m_median)/len(m_median)
m_median_avg
```

```python
w_median_avg = sum(w_median)/len(w_median)
w_median_avg
```

Subjects which are majority female don't offer higher salary.

# Use a grouped bar plot to compare the number of men with the number of women in each category of majors.

```python
recent_grads
```

```python
major_category = recent_grads.Major_category.unique()
print(major_category)
print(major_category.shape)
```

```python
fig, ax =plt.subplots()
x = range(16)
y = recent_grads.groupby("Major_category").Men.sum().values
ax.bar(x, y)
ax.set_xticks(x)
ax.set_xticklabels(major_category, rotation=90)
ax.set_xlabel("Majot_Category")
ax.set_ylabel("Numbers of Men")
```

```python
fig, ax =plt.subplots()
x = range(16)
y = recent_grads.groupby("Major_category").Women.sum().values
ax.bar(x, y)
ax.set_xticks(x)
ax.set_xticklabels(major_category, rotation=90)
ax.set_xlabel("Major Category")
ax.set_ylabel("Number of Women")
```

```python
fig, ax =plt.subplots()
x = range(16)
y = recent_grads.groupby("Major_category").Men.sum().values
yw = recent_grads.groupby("Major_category").Women.sum().values
ax.bar(x, y, label="Men")
ax.bar(x,yw, label="Women")
ax.set_xticks(x)
ax.set_xticklabels(major_category, rotation=90)
ax.set_xlabel("Majot_Category")
ax.set_ylabel("Numbers of Men")
ax.legend(loc="upper right")
```
